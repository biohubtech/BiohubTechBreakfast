```python
# Let's read a large TSV file containing chemical data
# We'll assign the "Name" column to be the row index
data = pandas.read_csv(open("/home/dwood/BiohubTech/properties.tsv"), delimiter="\t", index_col="Name")
data.head()
```

```python
# Which structures have a high molecular weight? >600?
data.loc[data.mw > 600].head()
```

```python
# Let's get rid of them; the squiggly line makes a boolean inversion, so we keep those with <600MW
print "Number of Rows Before", len(data)
data_cleaned = data.loc[~(data.mw > 600)]
print "Number of Rows After", len(data_cleaned)
```

```python
# Do any columns contain missing values?
print data_cleaned.isnull().head()
data_cleaned.isnull().apply(any, axis=0)
```

```python
# Which compounds contain the missing values?
data_cleaned.loc[data_cleaned.isnull().apply(any, axis=1)]
```

```python
# Let's get rid of those compounds
print "Number of Rows Before", len(data_cleaned)
data_cleaned = data_cleaned.loc[~data_cleaned.isnull().apply(any, axis=1)]
print "Number of Rows After", len(data_cleaned)
```

```python
# Nested sorts? Not a problem
data_cleaned.sort(columns=["hac", "mw"], ascending=[True, False]).head(10)
```

```python
# Now let's add some more data to our dataframe
# We'll assign the "Name" column to be the row index
risk_data = pandas.read_csv(open("/home/dwood/BiohubTech/risks.tsv"), delimiter="\t", index_col="Name")
data_cleaned = data_cleaned.merge(risk_data, left_index=True, right_index=True)
data_cleaned.head()
```

```python
# I want to calculate the % probability that each compound is safe
risk_columns = [ c for c in risk_data.columns if "Risk" in c  ]

# I convert the % to probabilities in range 0-1
# Then I invert to get the chance the compound is safe
# Then I multiply the columns together...
# convert back to risk (rather than safe)...
# and multiply by 100 to get %
risk_data["Overall Risk"] = (1 - (1 - (risk_data[risk_columns] / 100)).product(axis=1)) * 100
risk_data.head()
```
